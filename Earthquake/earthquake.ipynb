```python
## import the libraries

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
## load the earthquake dataset
df = pd.read_csv("all_month.csv")
df.head()
```

```python
## information on the dataset
df.info()
```

```python
df.describe()
```

```python
## checking fo missing values
df.isnull().sum()
```

```python
## plot histogram
df.hist(bins =10, figsize=(10,10))
plt.show()
```

Form the histogram, you can observe that the earthquakes happen to occur in a certain region of the world. Earthquakes seem to be more common in the positive latitude and the negative longtitude.

```python
## plot boxplot
df.boxplot(figsize=(20,30))
plt.show()
```

```python
## plot magnitude vs depth
plt.figure(figsize = (10, 7 ))
plt.scatter(
    x = df['mag'],
    y= df['depth'])
plt.xlabel("magnitude")
plt.ylabel("depth")
plt.title("Comparing magnitude and depth")
plt.show()
```

```python
## plot magnitude vs location source
plt.figure(figsize = (11,4))
plt. scatter(
    x = df["mag"],
    y = df["locationSource"])
plt.xlabel("magnitude")
plt.ylabel("location source")
plt.title("Magnitude vs location source")
plt.show()
```

```python
## scatter plot with mag, lat and long

x = df["latitude"]
y = df["longitude"]
z = df["mag"]

fig = plt.figure()
ax = np.array([fig.add_subplot(1,1,1, projection='3d')])

ax[0].scatter(x,y,z)
ax[0].set_title('comparing magnitude with the latitude and longtitude of the earthquakes')


plt.show()
```

The scatter plot graph shows that there is a  cluster of points in the region of high latitude and low longitude, whether it was a high magnitude or not. however, a very high concentrate of high magnitude earthqwuakes did seem to occur in the regions of high longitutdes.

```python
# simpler graph showing the location of the earthquakes


plt.figure(figsize = (11,4))
plt. scatter(
    x = df["longitude"],
    y = df["latitude"])
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.title("Location of the earthquakes in relation to the longitutde and latitude")
plt.show()
```

```python
##plot correlation matrix
corr_mat = df.corr()
corr_mat
```

```python
# heatmap

plt.figure(figsize=(10,10))
sns.heatmap(corr_mat, annot=True)
plt.title("correlation plot")
plt.show()
```

In a heatmap, we plot to see the relationship betweeen the two variables, the closer they are to either 1 or -1, they are more correlated to each other. Vice-versa, if the data is closer to 0 than it is to +/-1, that means they are not correlated to each other.
